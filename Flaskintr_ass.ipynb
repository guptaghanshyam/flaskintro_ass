Q1. What is Flask Framework? What are the advantages of Flask Framework?


Flask is a micro web framework for Python that allows you to build web applications. It is designed to be simple and lightweight, providing the essentials needed to create web applications without imposing too many constraints or libraries on the developer. Flask is widely used to develop small to medium-sized web applications and APIs.

Advantages of Flask Framework:

Lightweight: Flask is minimalistic and doesn't require a lot of boilerplate code. This makes it easy to learn and work with.

Flexible: Flask doesn't impose a particular way of doing things, giving developers the flexibility to choose components and libraries they prefer.

Modular: Flask is built as a collection of small components, allowing you to add or remove functionality as needed.

Extensible: You can easily extend Flask's functionality by adding third-party extensions.

Easy to Get Started: Setting up a basic Flask application is straightforward, making it great for beginners.

Well-Documented: Flask has comprehensive documentation, making it easier to understand and troubleshoot issues.

Good for Prototyping: Due to its simplicity, Flask is a good choice for quickly prototyping web applications.

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of defining URL routes for different parts of your web application. When a user makes a request to a specific URL, Flask uses these routes to determine which function (view) should be called to handle the request. App routing is crucial for mapping URLs to the appropriate code logic and generating responses.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used to generate URLs for routes defined in your application. It takes the name of the view function as an argument and returns the URL associated with that view. Here's an example:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return "Home Page"

@app.route('/about')
def about():
    return "About Page"

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home')) 
        print(url_for('about'))  
```
